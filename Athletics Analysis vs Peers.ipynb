```python
import pandas as pd
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
import matplotlib.pyplot as plt
```

```python
recruitingExpenses = pd.read_csv("OPE CSS Custom Data 2017-09-01 104534/Recruiting_Expenses_Men's_Women's_and_Coed_Teams_2003_2004_2005_2006_2007_2008_2009_2010_2011_2012_2013_2014_2015.csv")
```

```python
latest = recruitingExpenses[recruitingExpenses["Survey Year"] == 2015]
```

```python
pacU = latest[latest["Institution Name"] == "Pacific University"]
```

## Recruiting expenses are high vs national d3 w football

```python
fig = plt.figure(figsize=(12,6))
ax = latest['Men\'s Team Recruiting Expenses'].hist(bins=40)
ax.vlines(x=pacU['Men\'s Team Recruiting Expenses'], ymin=0, ymax=15, color='r')
plt.title("Men's Recruiting Expenses Histogram")
plt.xlabel("Expenses")
plt.ylabel("Count")
plt.savefig("mensrecruiting.png")
```

```python
fig = plt.figure(figsize=(12,6))
ax = latest['Women\'s Team Recruiting Expenses'].hist(bins=40)
ax.vlines(x=pacU['Women\'s Team Recruiting Expenses'], ymin=0, ymax=15, color='r')
plt.title("Women's Recruiting Expenses Histogram")
plt.xlabel("Expenses")
plt.ylabel("Count")
```
