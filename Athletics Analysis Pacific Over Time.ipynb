# Institutional Comparisons:

```python
comps = ["Pacific University", "George Fox University", "Linfield College","Willamette University"]
compdf = allDataComb[allDataComb['Institution Name'].isin(comps)]
```

```python
compdf.set_index('Survey Year', inplace=True)
compdf.groupby('Institution Name')["Men\'s Team Recruiting Expenses"].plot(legend=True)
```

## Recruiting expenses growing quickly

```python
recruitingROI = pd.DataFrame()
recruitingROI["Mens"] = compdf["Men\'s Team Recruiting Expenses"] / compdf["Unduplicated Count Men\'s Participation"]
recruitingROI["Womens"] = compdf["Women\'s Team Recruiting Expenses"] / compdf["Unduplicated Count Women\'s Participation"]
recruitingROI["Institution Name"] = compdf["Institution Name"]
```

```python
recruitingROI.groupby("Institution Name")["Mens"].plot(legend=True)
```

## Comparable-but-high $ per recruit for Men

```python
recruitingROI.groupby("Institution Name")["Womens"].plot(legend=True)
```

## But why is it so hard to recruit women to participate in PacU athletics?
